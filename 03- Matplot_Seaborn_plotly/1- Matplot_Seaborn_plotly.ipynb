## Creating Multiplots on Same Canvas

### Matplot and Seaborn

```python
# Sample data for Matplotlib
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Sample data for Seaborn
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5] * 3,
    'y': [2, 3, 5, 7, 11, 1, 4, 6, 8, 10, 2, 5, 7, 9, 12],
    'category': ['A'] * 5 + ['B'] * 5 + ['C'] * 5
})
```

```python
# Create a figure for multiple plots
plt.figure(figsize=(12, 6))

# Matplotlib subplot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Prime Numbers', color='b', linestyle='-', linewidth=2, marker='o')
plt.title('Matplotlib Plot 1')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True, linewidth=0.6, alpha=0.3)

# Matplotlib subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Random Numbers', color='r', linestyle='-', linewidth=2, marker='s')
plt.title('Matplotlib Plot 2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True, linewidth=0.6, alpha=0.3)

# Seaborn subplot 1
plt.subplot(2, 2, 3)
sns.lineplot(data=data[data['category'] == 'A'], x='x', y='y', marker='o')
plt.title('Seaborn Plot 1')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True, linewidth=0.6, alpha=0.3)

# Seaborn subplot 2
plt.subplot(2, 2, 4)
sns.lineplot(data=data, x='x', y='y', hue='category', style='category', markers=True, dashes=False, palette='viridis')
plt.title('Seaborn Plot 2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True, linewidth=0.6, alpha=0.3)


# Adjust layout and display plot
plt.tight_layout()
plt.show()
```

```python
courses =['C','C++','Java', 'Python']
values  =[10 , 20  , 30   ,  40]

fig = plt.figure(figsize = (10, 5))
plt.bar(courses, values, color ='maroon', width = 0.4,alpha=0.95)

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.grid(True, linewidth=0.7, alpha=0.5)
plt.show()
```

```python
# Sample data
categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = [4, 7, 1, 8, 5, 9]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd' ,'#6377bd']
hatches = ['/', '\\', '|', '-', '+' ,'o']
```

```python
# Create figure and axis
plt.figure(figsize=(10, 6))

# Plot data
bars = plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5, hatch=hatches)

# Customize plot
plt.title('Customized Matplotlib Bar Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True, linewidth=0.8, alpha=0.5)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height - 0.3, f'{height}', ha='center', va='bottom', color='white')

# Add legend
plt.legend(bars, categories, title='Categories')

# Display plot
plt.show()
```

## Multiple bar plots

### Matplot

```python
# Data
data = [
    [5, 10, 15],
    [7, 14, 21]
]
labels = ['Group 1', 'Group 2']
colors = ['blue', 'green']
tick_labels = ['A', 'B', 'C']
```

```python
# Plot using Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
num_datasets = len(data)
bar_width = 0.8 / num_datasets
index = range(len(data[0]))

for i, dataset in enumerate(data):
    bar_position = [j + i * bar_width for j in index]
    bars = ax.bar(bar_position, dataset, bar_width, label=labels[i] if labels else f'Dataset {i+1}', color=colors[i] if colors else None)
    
    # Adding data labels
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')

ax.set_title('Multiple Bar Plots')
ax.set_xlabel('Category')
ax.set_ylabel('Values')
ax.set_xticks([r + bar_width for r in range(len(data[0]))])
ax.set_xticklabels(tick_labels)
ax.legend(loc='upper left')
ax.grid(True, linewidth=0.5, alpha=0.3)

plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values1 = [3, 7, 5, 2]
values2 = [4, 6, 2, 5]
x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars
```

```python
fig, ax = plt.subplots()

# Plotting the first set of bars
rects1 = ax.bar(x - width/2, values1, width, label='Group 1')

# Plotting the second set of bars
rects2 = ax.bar(x + width/2, values2, width, label='Group 2')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Category')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and category')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()
# Add gridlines
ax.grid(True, linewidth=0.8, alpha=0.5)

# Add annotations
for rect in rects1 + rects2:
    height = rect.get_height()
    ax.annotate('{}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# Adjust layout to make room for the annotations
fig.tight_layout()
plt.show()
```

### Seaborn

```python
data = [
    [5, 10, 15],
    [7, 14, 21]
]
df = pd.DataFrame(data).T
labels = ['Group 1', 'Group 2']
colors = ['blue', 'green']
tick_labels = ['A', 'B', 'C']
```

```python
df.columns = labels if labels else [f'Dataset {i+1}' for i in range(len(data))]
df['Category'] = tick_labels
df = df.melt(id_vars='Category', var_name='Dataset', value_name='Value')

# Plot using Seaborn
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df, x='Category', y='Value', hue='Dataset', palette=colors if colors else None)

# Adding data labels
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')

plt.title('Multiple Bar Plots')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend(loc='upper left')
plt.grid(True, linewidth=0.8, alpha=0.5)

plt.show()
```

# Bie_chart

### Matplot and Seaborn

```python
cars = ['AUDI', 'BMW', 'FORD','TESLA', 'JAGUAR', 'MERCEDES']

data = [16, 12, 19, 21, 8, 24]

explode = (0.1, 0.0, 0.2, 0.3 , 0.0 ,0.0)

colors = ( "orange", "cyan", "brown","grey", "indigo", "beige")

wp = {'linewidth' : 1, 'edgecolor' : 'green'}

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax. pie(data, autopct=lambda pct: func(pct,data), explode=explode, labels=cars, shadow=True, colors=colors, startangle=90,wedgeprops= wp, textprops=dict(color = 'magenta'))

ax.legend(wedges, cars, title='Cars', loc='upper right', bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size= 8, weight = 'bold')

ax.set_title("Customizing pie chart")

plt.show()
```

```python
# Create a figure and axis
plt.figure(figsize=(14, 8))

# Create a heatmap with customization
heatmap = sns.heatmap(
    data=df,
    annot=True,              # Annotate cells with the data value
    fmt=".2f",               # Format the annotations to 2 decimal places
    cmap="YlGnBu",           # Colormap
    cbar=True,               # Show color bar
    cbar_kws={'label': 'Scale'},  # Color bar customization
    linewidths=0.5,          # Line width between cells
    linecolor='gray',        # Line color between cells
    square=True,             # Force square cells
    mask=df<0.1,               # Do not mask any data
    annot_kws={"size": 10},  # Annotation font size
    xticklabels=True,        # Show x-axis labels
    yticklabels=True,        # Show y-axis labels
    robust=True              # Robust colormap limits
)

# Customize the plot
plt.title('Customized Seaborn Heatmap', fontsize=18, weight='bold', pad=20)
plt.xlabel('X', fontsize=14, weight='bold')
plt.ylabel('Y', fontsize=14, weight='bold')
# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')
# Rotate y-axis labels
plt.yticks(rotation=0)
# Adjust layout for better fit
plt.tight_layout()
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.5)
# Display the heatmap
plt.show()

```

# contour

### Matplot

```python
x = np.linspace(-5, 5, 40)
y = np.linspace(-5, 5, 40)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
plt.contour(X, Y, Z)
plt.title('Contour Plot')
plt.show()
```
